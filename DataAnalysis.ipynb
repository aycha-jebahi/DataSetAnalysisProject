```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('diamonds_dataset.csv')
```

```python
sns.relplot(x='carat', y='price', data=df)
plt.title('Prices of Carats\n', fontsize=16)
plt.tight_layout()
```

```python
sns.relplot(x='color', y='price', data=df)
plt.title('Prices of Colors\n', fontsize=16)
plt.tight_layout()
```

```python
# Prices of Carats of each Shape
sns.relplot(x='carat', y='price', data=df, col='shape' , col_wrap=3, height=4)
plt.tight_layout()
```

```python
# Prices of Carats of each Color
sns.relplot(x='carat', y='price', data=df, col='color', col_wrap=3, height=4)
plt.tight_layout()
```

```python
# Prices of Carats of each Cut
sns.relplot(x='carat', y='price', data=df, col='cut' , col_wrap=3, height=4)
plt.tight_layout()
```

```python
# Prices of Carats of each Clarity
sns.relplot(x='carat', y='price', data=df, col='clarity' , col_wrap=3, height=4)
plt.tight_layout()
```

```python
# Prices of Carats of Natural and Lab Diamonds
sns.relplot(x='carat', y='price', data=df, col='type')
plt.tight_layout()
```

```python
# Prices of Carats of each shape For Natural and Lab Diamonds
sns.relplot(x='carat', y='price', data=df, kind='line', hue='type', palette=['#04ff21', '#fe3a9e'], col='shape', col_wrap=3, height=4)
plt.grid(alpha=0.5)
```

```python
# Prices of Carats of each Cut For Natural and Lab Diamonds
sns.relplot(x='carat', y='price', data=df, kind='line', hue='type', palette=['#04ff21', '#fe3a9e'], col='cut' , col_wrap=3, height=4)
plt.grid()
```

```python
sns.relplot(x='color', y='price', data=df, kind='line', hue='carat' , col='type')
plt.title('Prices of Carats of Natural and Lab Diamond Colors\n', fontsize=16)
plt.grid()
```

```python
sns.relplot(x='clarity', y='price', data=df, kind='line', hue='carat' , col='type')
plt.title('Prices of Carats of Natural and Lab Diamond Clarities\n', fontsize=16)
plt.grid()
```

```python
sns.relplot(x='cut', y='price', data=df, kind='line', hue='carat' , col='type')
plt.title('Prices of Carats of Natural and Lab Diamond Cuts\n', fontsize=16)
plt.grid(alpha=0.9)
```

```python
sns.catplot(x='type', data=df, kind='count')
plt.title('Count of Each Diamond Type\n', fontsize=16)
plt.tight_layout()
```

```python
sns.catplot(x='report', data=df, kind='count')
plt.title('Count of Each Diamond Report\n', fontsize=16)
plt.tight_layout()
```
